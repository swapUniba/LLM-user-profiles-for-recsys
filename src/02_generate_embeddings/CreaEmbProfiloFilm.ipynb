```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import pickle
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')

#File da cui leggere le preferenze utente
directory_path = "/content/drive/MyDrive/ml1m"
```

```python
# Carica i profili utente tramite pickle
profiles = pickle.load(open("/content/drive/MyDrive/ml1m/dizionario.pkl", 'rb'))

# Inizializza il modello di encoding
model = SentenceTransformer('all-MiniLM-L12-v1')

# Esempio per ottenere la dimensione degli embedding
text = 'this is a test'
text_emb = model.encode(text)
emb_dim = len(text_emb)

# Salva gli embedding del testo
embeddings = dict()

```

```python
# Iterazione sugli utenti per la creazione degli embedding
for user in profiles:
    profile = profiles[user]

    # Verifica se il profilo è un dizionario e contiene la chiave 'plain'
    if isinstance(profile, dict) and 'plain' in profile:
        profile_text = profile['plain'].strip()  # Estrai il testo e rimuovi spazi vuoti
    else:
        print(f"Profilo non valido per l'utente {user}. Non è presente il campo 'plain'.")
        profile_text = ""

    if not profile_text:  # Se il testo del profilo è vuoto
        print(f"Profilo vuoto per l'utente {user}. Creazione di un embedding vuoto.")
        embedding = np.zeros(emb_dim)  # Array di zeri come fallback
    else:
        try:
            embedding = model.encode(profile_text)
            print(f"Embedding creato correttamente per l'utente {user}")
        except Exception as e:  # Cattura l'eccezione e stampa un messaggio di errore
            print(f"Error encoding profile for user {user}: {e}")
            embedding = np.zeros(emb_dim)  # Array di zeri con la dimensione corretta

    embeddings[user] = embedding

# Salva gli embedding su un file
with open('/content/drive/MyDrive/ml1m/embProfiliUtentiFilm.pickle', 'wb') as f:
    pickle.dump(embeddings, f)

print("Embedding saved successfully.")
```

```python
#Estendiamo il file user-profile_item_emb.pkl
# Carica l'attuale file user-profile_item_emb.pkl
with open('/content/drive/MyDrive/ml1m_emb/exp_embs/user-profile_item_emb.pkl', 'rb') as f:
    item_embeddings = pickle.load(f)

# Estendi il file con gli embedding degli utenti
item_embeddings.update(embeddings)

# Salva il dizionario esteso
with open('/content/drive/MyDrive/ml1m_emb/exp_embs/user-profile_item_emb.pkl', 'wb') as f:
    pickle.dump(item_embeddings, f)

print(f"File aggiornato con successo.")
```
