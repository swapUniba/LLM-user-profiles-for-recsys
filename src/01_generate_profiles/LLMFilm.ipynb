```python
import os

first_user = 470
last_user = 470

# vedo quali sono tutti gli utenti del dataset
all_users = set(sorted(ratings[0]))

# controllo il primo utente per il quale non abbiamo il profilo
#for user in all_users:

  # appena lo trovo fermo la ricerca
 # if not os.path.exists(f'/content/drive/MyDrive/ml1m/outputs/user{user}_sysprompt2.json'):
  #  first_user = user
  #  break

# costruisco il profilo di 20 utenti alla volta - possiamo aumentare questo numero
last_user = first_user


# per ogni utente
for user_i in range(first_user,last_user+1):

  # leggo i rating positivi
  start_time = time.time()
  user_i_pos_ratings = pos_ratings.loc[pos_ratings[0]==user_i]
  user_i_pos_ratings_size = int(user_i_pos_ratings.size/3)
  print("Starting user ", user_i, "... (", user_i_pos_ratings_size, " positive ratings )")

  # leggo i nomi degli item che piacciono all'utente e costruisco il prompt
  prompt = "My favourite movies are: "
  for i in range(user_i_pos_ratings_size):
    mov = (names.loc[names[0]==user_i_pos_ratings[1].values[i]])[1].values[0]
    mov = mov.replace("http://dbpedia.org/resource/","")
    mov = mov.replace("_"," ")
    mov = re.sub("[\(\[].*?[\)\]]", "", mov).strip()
    prompt = prompt + mov + "; "
  prompt = prompt[:-2] + "."

  # richiesta al LLM
  input_text = tokenizer.apply_chat_template([{'role': 'user', 'content': prompt}, {'role': 'system', 'content': sysprompt2}], tokenize=False)

  # prendo il risultato
  result = pipe(input_text)[0]['generated_text']
  print(result)

  # dict per il risultato
  mydict = {"plain": result}

  # formatto meglio il risultato
  result = result.replace('<','')
  result = result.replace('>','')
  result = result.split('.')

  # controllo che il risultato sia corretto
  broken = False
  for line in result:
    if (len(line.strip().split(': ')) >= 2):
      feat = line.replace("Favourite ","").strip().split(': ')[0].lower()
      vals = line.strip().split(': ')[1].lower()
      valsList = vals.split(', ')

      if len(valsList) > 11:
        print("BROKEN\n")
        print(valsList)
        print("\nBROKEN")
        broken = True
        break
      mydict[feat] = valsList

  json_object = json.dumps(mydict)

  # se il risultato è corretto, lo salvo nella cartella
  if not broken:
    print("JSON file created correctly.")
    filename = "/content/drive/MyDrive/ml1m/outputs/user" + str(user_i) + "_sysprompt2.json"
    with open(filename, "w") as outfile:
        outfile.write(json_object)

  # altrimenti salvo il fatto che si è rotto - ci servirà in seguito
  else:
    filename = "/content/drive/MyDrive/ml1m/broken_outputs/user" + str(user_i) + "_sysprompt2.json"
    filename = "/content/drive/MyDrive/ml1m/broken/user" + str(user_i) + "_sysprompt2.json"
    with open(filename, "w") as outfile:
        outfile.write(json_object)
    with open("/content/drive/MyDrive/ml1m/broken.txt", "a") as brokenfile:
      brokenfile.write(str(user_i)+" ")
      brokenfile.close()


  end_time = time.time()
  print("User ", user_i, " finished in ", round(end_time-start_time,2), " seconds.\n")
```
